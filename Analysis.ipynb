```python
%matplotlib

import matplotlib.pyplot as plt
import numpy as np
import math

tt= np.loadtxt("/home/lucifer/IISER/Summer_2021/Energy_Eigenvalues_SingleLayer_Chirality_Average/Result_t[1.50]")
#print(tt)
i=0
result=[]
while i<len(tt):
    #print(result1[i])
    if (tt[i]>=0 and tt[i]<1):
        #print(tt[i])
        result.append(tt[i])
    i=i+1
#print(result)

result = [i for i in result if i != 0]
#print(result)

i=0
rt=[]
while i<len(result):
    #x = -math.log(math.log(1/result[i])) #Dyson form
    x = -(math.log(1/result[i])) #Gade wegner
    rt.append(x)
    i=i+1
#print(rt)

#print(result)

i=0
xx=[]
d=0
num = 20
while i<num:
    d = d+ 0.03
    xx.append(d)
    i=i+1
p=sorted(xx)
#print(p)

j=0
yy=[]
while j<len(p)-1:
    d1 = p[j]
    d2 = p[j+1] #np.log(4+np.log(j))
    #print(d1,d2)
    #print(d1,d2)
    d = len([x for x in result if x >= d1 and x <= d2])
    #print(d)
    dd= d1*d 
    #print(dd)
    ddd = np.log(dd)
    #print("**")
    #print(ddd)
    yy.append(ddd)
    j=j+1
#print(yy)
f=rt[:19]
#print(f)

#i=1
#while i<4:
#    f.pop(i)
#    yy.pop(i)
#    i=i+1
#del f[1:4]
#del yy[1:4]
#print(f)
plt.xlabel('-ln(ln(1/E))')
plt.ylabel(r'$ln(E \rho (E))$')
plt.title("Density of States, with no next neighbour hopping, w = 0.50, sample size 3200")
#plt.scatter(f,yy,s=5,c="black")
#plt.plot(f,yy)
#plt.show()



# calculate polynomial
z = np.polyfit(f, yy, 1)
fd = np.poly1d(z)

# calculate new x's and y's
x_new = np.linspace(f[0], f[-1], 50)
y_new = fd(x_new)

plt.plot(f,yy,'x', x_new, y_new,markersize=5)
#plt.errorbar(f, yy, yerr=error_array)

#plt.xlim([f[0]-1, f[-1] + 1 ])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import RadioButtons

t = np.arange(0.0, 2.0, 0.01)
s0 = np.sin(2*np.pi*t)
s1 = np.sin(4*np.pi*t)
s2 = np.sin(8*np.pi*t)

fig, ax = plt.subplots()
l, = ax.plot(t, s0, lw=2, color='red')
plt.subplots_adjust(left=0.3)

axcolor = 'lightgoldenrodyellow'
rax = plt.axes([0.05, 0.7, 0.15, 0.15], facecolor=axcolor)
radio = RadioButtons(rax, ('2 Hz', '4 Hz', '8 Hz'))


def hzfunc(label):
    hzdict = {'2 Hz': s0, '4 Hz': s1, '8 Hz': s2}
    ydata = hzdict[label]
    l.set_ydata(ydata)
    plt.draw()
radio.on_clicked(hzfunc)


plt.show()
```
